```python
import pandas as pd #imports panda library
import numpy as np
import seaborn as sns #imports seaborn library

# Matplot
import matplotlib.pyplot as plt #import matplot library
%matplotlib inline
```

```python
df=pd.read_csv('final_movies_list.csv') #Imports my csv file from my jupyter notbook
```

```python
df.head() #This will give you the first 5 of 50 listings
```

```python
df.info() #DataFrame Info 
```

```python
df['gross'] = df['gross'].str.replace(',', '') #replacing comma 
df['gross'] = df['gross'].str.replace('$', '') #replace $
df['gross'] = df['gross'].astype('int64') #turning it into a number
```

```python
ax = sns.countplot(x="maturity_rating",data=df, palette="Set3", order = df['maturity_rating'].value_counts().index)
ax.set_title('Genre Frequency')
ax.set_ylabel('Frequency')
ax.set_ylabel('Maturity') 
#Using Countplot to graph maturity raitings
```

```python
#ax = sns.barplot(y="titles",x='gross', data=df.head(), palette="Set3") 
#ax.set_title('Top Most Grossed Movies Of All Time')
#ax.set_ylabel('$ Amount')
#ax.set_ylabel('titles')
```

```python
sns.set_style('whitegrid') 
ax = sns.violinplot(y="titles",x='gross', data=df.head())
ax.set_title('Top Most Grossed Movies Of All Time')
ax.set_ylabel('$ Amount')
ax.set_ylabel('titles') #Here we used violinplot to graph top most grossed movies of all time
```

```python
sns.swarmplot(x='gross', y='genre', data=df) #Here, we used Swarmplot to graph our top grossing genres
```

```python
ax = df['genre'].value_counts().plot(kind='bar', figsize=(14,8),title="Genre Frequency")
ax.set_xlabel("Genre Type")
ax.set_ylabel("Frequency") #using a bargraph to find the most watched genres
```
