```python
import pandas as pd #pandas library
import requests
from bs4 import BeautifulSoup #BeautifulSoup Library
```

```python
page = requests.get('https://www.imdb.com/list/ls000021718/') #Pulling the URL from IMBD
soup = BeautifulSoup(page.content,'html.parser') #Getting content from source
top_movies= soup.find(id="main") #browsing thru the page
```

```python
movies = [] #List of the movies
all_movies = soup.find_all('div', {'class':'lister-item mode-detail'}) 
#Populating the list of the top 50 movies
for movie in all_movies: 
    #looping for individual movies inside my movie list
    ids = movie.find_all('div', {'class' : "lister-item-image ribbonize"}) 
    ids = str(ids).split('=')[2].split('>')[0].strip('"') 
    #formating raw id from source by using the split function
    title = movie.find('div', {'class': 'lister-item-content'})
    title = str(title)
    title = title.split('>')
    title = title[5]
    title = title.split('<')
    title = title[0]
    maturity_rating = movie.find(class_="certificate").get_text() 
    #Assigning certificate from source to maturity_rating
    genre = movie.find("span", {"class":"genre"}).text.strip() 
    result = (ids, title, maturity_rating, genre) 
    #We now have our columns
    movies.append(result)     
      
 
movies_df = pd.DataFrame(movies, columns = ['ids', 'titles', 'maturity_rating', 'genre']) #Assigning the code a name
movies_df
```

```python
def getGross(movie_id): #getting the $ amount
    url = 'https://www.imdb.com/title/' + movie_id + '/?ref_=ttls_li_tt' 
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'html.parser')
 
    main = soup.find(id="titleDetails")
    gross = main.find_all('div', {'class':'txt-block'})
    if '$' in gross[9].text.strip().split(':')[1]: 
        #if there is a $ in gross, use the 10th spot else... 
        gross = gross[9].text.strip().split(':')[1]
    else:
        gross = gross[8].text.strip().split(':')[1] 
        #If the $ is not in gross, go to the 9th spot 
    return (movie_id, gross) 
    #return the gross 
 
moviegross = [] #declaring array list
 
for movie in movies_df['ids']: 
    #for each movie in the movie_df list, get gross info 
    moviegross.append(getGross(movie))
```

```python
df = pd.DataFrame(moviegross, columns = ['ids', 'gross']) #giving it a name
final_movie_list = movies_df.merge(df, on = 'ids') #giving it a name 
```
